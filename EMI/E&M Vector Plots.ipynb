```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

rc('text.latex',preamble='\\usepackage{libertine}\n\\usepackage[libertine]{newtxmath}')
rc('font',**{'family':'serif','serif':['Linux Libertine O']}, size=18)
rc('text', usetex=True)
```

```python
# to make your own vector field plot, change the below function to match your x and y components

def vx(x, y):
    return x**2

def vy(x, y):
    return -y**2

def plot_vector_field_2d(vx, vy):
    x = np.linspace(-2, 2, 11)
    y = np.linspace(-2, 2, 11)
    x, y = np.meshgrid(x, y)

    vvx = vx(x, y)
    vvy = vy(x, y)
    

    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(1,1,1)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_aspect('equal')
    ax.quiver(x, y, vvx, vvy, pivot='middle')


plot_vector_field_2d(vx, vy)

```

```python
# some examples of divergences

plot_vector_field_2d(lambda x, y: x, lambda x,y: y)
plot_vector_field_2d(lambda x, y: 0, lambda x,y: 1)
plot_vector_field_2d(lambda x, y: 0, lambda x,y: y)

# some curls

plot_vector_field_2d(lambda x, y: -y, lambda x,y: x)
plot_vector_field_2d(lambda x, y: 0, lambda x,y: x)
```
