## A Massive String

In class we saw that for the hangining block problem with a massless string, the acceleration of the blocks was given by
$$
y_\text{massless}(t) = -\frac{1}{4} gt^2.
$$
For the massive string case, though, it's the more complicated
$$
y_\text{massive}(t) = \frac{m}{m_s} L [1 - \cosh (\sqrt{\alpha} t)],
$$
where $\alpha = m_s g / L (2m+m_s)$.

To see the difference (and similarity) between these, let's plot them up.  I'll pick $L = 1.0$ m, $m = 1.0$ kg, and $m_s = 0.5$ kg.

```python
%matplotlib inline

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc

rc('text', usetex=True)
matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams.update({'font.size':14})
plt.rc('axes', labelsize=16)
plt.rcParams.update({'figure.autolayout': True})

g = 9.8
L = 1.0
m = 1.0
m_s = 1.0
alpha = m_s * g / L / (2.0*m + m_s)

# first we'll generate some arrays for our two equations
t = np.arange(0, 1.0, 0.01)
y_massless = -0.25 * g * t**2
y_massive = m * L / m_s * (1 - np.cosh(np.sqrt(alpha) * t))

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(2,1,1)
ax.plot(t, y_massless, color='black')
ax.plot(t, y_massive, color='red')
ax.set_xlim(0.0, 0.7)
ax.set_ylim(-1.0, 0.05)
ax.set_xlabel(r'$t$ (s)')
ax.set_ylabel(r'$y$ (m)')

ax = fig.add_subplot(2,1,2)
ax.plot(t, y_massive-y_massless, color="black")
ax.set_xlim(0.0, 0.7)
ax.set_xlabel(r'$t$ (s)')
ax.set_ylabel(r'$\Delta y$ (m)')

plt.savefig("massive_string.pdf")
```
